## Teorema Central do Limite

Este notebook possui como fim ilustrar o **teorema central do limite**.

Segundo este teorema: 

→ Dado um subconjunto de amostras obtidas a partir de uma população, pode-se calcular a média dos seus valores. Quando esse procedimento é executado com outros subconjuntos, de forma repetida e independente, torna-se possível construir uma distribuição das médias (sampling distribution).

Essa distribuição resultante irá se aproximar à uma distribuição normal.

Isso, mesmo que a distribuição da população original não seja uma distribuição normal.

```python
# Bibliotecas necessárias

import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# Definindo o estilo dos gráficos

plt.style.use('seaborn')
```

```python
# Variáveis inicias

random_numbers = np.random.randint(101, size = 1000)

bins = np.linspace(0, 100, num = 100)
```

```python
# Histograma dos números aleatórios

plt.hist(random_numbers, bins = bins)

plt.grid(False)

plt.title('Histograma - Números Aleatórios')

plt.xlabel('Intervalo numerico')

plt.ylabel('Número de amostras');
```

```python
# Construindo uma sampling distribution da média de subconjuntos de amostras

mean_samples = []

for _ in range(100_000):
    
    sample = np.random.choice(random_numbers, size = 100)
    
    mean_sample = np.mean(sample)
    
    mean_samples.append(mean_sample)
```

```python
# Obtendo métricas da distribuição normal

std = np.std(mean_samples)

mean = np.mean(mean_samples)

interval_min = int(mean - 4 * std)

interval_max = int(mean + 4 * std)
```

```python
# Histograma da sampling distribution

plt.hist(mean_samples, bins = bins[interval_min : interval_max])

plt.grid(False)

plt.title('Histograma - Sampling Distribution')

plt.xlabel('Média do subconjunto')

plt.ylabel('Número de amostras');
```
